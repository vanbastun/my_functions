```python
def my_binare_list_search(data, n):
    """
    
    По списку data и числу n возвращает минимальное ближайшее к n число, через бинарный поиск
    
    """
    temp_list = data.copy()
    while len(temp_list) > 1:
        middle_i = len(temp_list) // 2
        if n > temp_list[middle_i]:
            temp_list = temp_list[middle_i:]
        else:
            temp_list = temp_list[:middle_i]
    return temp_list[0]
```
